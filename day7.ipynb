```python
#generate a random numbers between 45 and 80 total should be 200
# modify this numbers by finding there square root
# filter out the odd and the even numbers
```

```python
import random
import math
numbers = []
counter = 1
while counter <= 200:
    num = random.randint(45,80)
    numbers.append(num)
    counter += 1


updatenumber1 = list(map(lambda x:math.sqrt(x),numbers))
print(updatenumber1)
odd = list(filter(lambda x: x%2 == 1, numbers))
even = list(filter(lambda x: x%2 == 0, numbers))
print(odd)
print('\n')
print(even)
```
